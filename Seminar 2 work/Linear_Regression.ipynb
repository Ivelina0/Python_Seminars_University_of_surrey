**Mean Squared Error** (MSE) is the mean of the squared errors:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

```python
print('Mean Square Error', metrics.mean_squared_error(y_test, prediction))
```

**The coefficient of determination, denoted R2** is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

The Residual Sum of Squares: $$ SS_{res} = {\sum_{i=1}^n(y_i-\hat{y}_i)^2} $$
or $$ \ SS_{res} = {\sum_{i=1}^n(y_i-(\alpha + \beta x_i))^2} $$

And the total sum of squares: $$ SS_{tot} =  {\sum_{i=1}^n(y_i-\bar{y})^2} $$

and the R-squared metric is: $$ R^2 = 1 - \frac {SS_{res}}{SS_{tot}} $$

```python
print('Model R^2 Square value', metrics.r2_score(y_test, prediction))
```

```python
# Model Regression line on test set

plt.scatter(x_test, y_test, color='darkblue')

plt.plot(x_test, prediction,':', color='black', linewidth=3)

plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Linear Regression')
```

# Linear Regression using normal equations

Here is an example from Andrew Ng's course on Coursera: 

```python
from IPython.display import Image
Image("Linear_reg_image.png")
```

```python
# Define a function that calculates theta
def theta_calc(x_train, y_train):
    
    #Initializing all variables
    n_data = x_train.shape[0] # finding the amount of rows in the x_train series
    bias = np.ones((n_data,1)) # creating an array with 1s of the same size as the amount of rows we found
    x_train_b = np.append(bias, x_train, axis=1) # merging the two series into a 2D arrays
    
    # 
    theta_1 = np.linalg.inv(np.dot(x_train_b.T,x_train_b)) # inverting the matrix
    theta_2 = np.dot(theta_1, x_train_b.T) # dot product with transpose of x
    theta = np.dot(theta_2,y_train) # dot product with the y array
    #
    return theta
```

```python

def predict_func(slope,intercept,x_test):
    #
    pred = []
    n_data = x_test.shape[0] # Note we have 20 test sets
    
    for i in range(n_data): # create a list of the y hat values by multiplying: the slope * x + the intercept 
        
        pred.append((slope * x_test[i]) + intercept)
        
    return pred
```

```python
def mse_calc(prediction, y_test):
    #
    total_data = len(prediction) # find n: the number of data sets
    
    error = 0 # initiate an error of 0 
    
    # Mean Squared Error
    error = (np.sum((prediction - y_test)**2))/total_data # update the error value
    
    return error
```

```python
def rsq(prediction, y_test):
    
    total_data = len(prediction)
    
    #Average of total prediction 
    y_avg = np.sum(y_test)/total_data
    
    #total sum of square error
    tot_err = np.sum((y_test-y_avg)**2)
    
    #total sum of squared error of residuals
    res_err = np.sum((y_test-prediction)**2)
    
    r2 = 1 - (res_err / tot_err)
    return r2
```

```python
#Finding optimal theta value using normal equations

theta = theta_calc(x_train, y_train)

intercept = theta[0]

slope = theta[1]

print('Intercept of the model', intercept)
print('Slope of the model', slope)
```

```python
#Prediction calculation
prediction = predict_func(slope, intercept, x_test)
```

```python
#MSE calculation
error =  mse_calc(prediction, y_test)
print('Mean squared error of the model', error)

#R-square calculation
r2_val = rsq(prediction, y_test)
print('R squared value', r2_val)
```

```python
#Residual plot
plt.scatter(prediction, prediction - y_test, c='g', s = 40)

plt.hlines(y=0, xmin=0, xmax=100)

plt.title('Residual plot')
plt.ylabel('Residual');
```

# Linear Regression using Gradient Descent

Here is a good article that explains the gradient descent algorithm to finding the optimal thata: https://towardsdatascience.com/linear-regression-simplified-ordinary-least-square-vs-gradient-descent-48145de2cf76

But the basic steps are:
1. Find the difference between the actual y and predicted y value(y = mx + c), for a given x.
2. Square this difference.
3. Find the mean of the squares for every value in X

We define our hypothesis function $$ h_\theta(x) = \theta_0 + \theta_1x $$ for which we want to find the parameters theta0 and theta1.

From the hypothesis function we can find the cost function $$ J(\theta_0, \theta_1) = \frac 1{2m}\sum(h_\theta(x_i) - y_i)^2  $$ which we want to minimise. It is just the sum of squares of the predicted values minus the true ones. Similar to the Mean Squared Error function.

We then take the partial derivatives of the cost function with respect to the two parameters theta0 and theta1:

$$ \theta_0 = \frac{\partial}{\partial \theta_0}J(\theta_0 \theta_1) = \frac 1{m}\sum(h_\theta(x_i) - y_i) $$

$$ \theta_1 = \frac{\partial}{\partial \theta_1}J(\theta_0 \theta_1) = \frac 1{m}\sum(h_\theta(x_i) - y_i)*x_i $$

To find the minimum of the cost function h(x), repeat the steps to apply various values for theta 0 and theta 1. In other words, repeat steps until convergence. Choose a learning rate alpha you think fits well.

$$ \theta_0 = \theta_0 - \alpha \frac 1{m}\sum(h_\theta(x_i) - y_i) $$

$$ \theta_1 = \theta_1 - \alpha \frac 1{m}\sum(h_\theta(x_i) - y_i)*x_i $$

The code for a basic Gradient Descent algorithm:

```python
def mse_calc(slope, intercept, x_train, y_train):
    
    tot_error = 0
    pred = []
    
    for i in range(len(x_train)):
        
        # calculating total error. It follows the formula y=mx+c
        # m is the slope and c is the intercept
        
        pred.append(((slope*x_train[i]) + intercept))
        
        tot_error += (y_train[i] - ((slope*x_train[i]) + intercept)) ** 2
        
    error = tot_error / float(len(x_train))
    return error, pred
```

```python
# gradient descent 
# We can pick a learning rate l_rate and the number of iterations we want iter_val

def grad_descent(s_slope, s_intercept, l_rate, iter_val, x_train, y_train):
    
    for i in range(iter_val):
        int_slope = 0
        int_intercept = 0
        n_pt = float(len(x_train))
        
        
        for i in range(len(x_train)):
            int_intercept += - (2/n_pt) * (y_train[i] - ((s_slope * x_train[i]) + s_intercept))
            int_slope += - (2/n_pt) * x_train[i] * (y_train[i] - ((s_slope * x_train[i]) + s_intercept))
            
        final_slope = s_slope - (l_rate * int_slope)
        
        final_intercept = s_intercept - (l_rate * int_intercept)
        
        s_slope = final_slope
        s_intercept = final_intercept
        
        
    return  s_slope, s_intercept
```

```python
# defining the initial parameters: slope and intercept value as 0 

learning_rate = 0.0001

start_slope = 0

start_intercept = 0

iteration = 50
```

```python
# Intial run using the functions above:

grad_slope, grad_intercept = grad_descent(start_slope, start_intercept, learning_rate, iteration, x_train, y_train)

final_e_value, prediction = mse_calc(grad_slope, grad_intercept, x_test, y_test)

print('Slope of the model', grad_slope)
print('Intercept of the model', grad_intercept)

print('Error value of the model', final_e_value)

# From the rsq function we defined in the Normal Equations example

r2_val = rsq(prediction, y_test)
print('R squared value', r2_val)
```

```python
#Graph
plt.scatter(x_test, y_test)
plt.plot(x_test, prediction, color='blue', linewidth = 3)
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
